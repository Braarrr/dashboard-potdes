```python
import pandas as pd

# Baca layout
layout = pd.read_excel("podes_fokus_padat.xlsx")

# Mapping kode → label
rename_dict = dict(zip(layout["Name"].str.lower(), layout["Label_Short"]))

# Samain case kolom dataset
df.columns = df.columns.str.lower()

# Rename kolom pakai layout
df = df.rename(columns=rename_dict)

# Bersihin nama kolom: 
#  - lowercase
#  - ganti spasi dengan underscore
#  - hapus karakter non-alfanumerik
df.columns = (
    df.columns.str.lower()
              .str.replace(" ", "_")
              .str.replace(r"[^\w]", "", regex=True)  # buang karakter aneh
)

# Simpan hasil
df.to_csv("dataset_r7_clean.csv", index=False)

print("Kolom sudah diganti nama deskriptif tanpa spasi.")

```

```python
import pandas as pd
df = pd.read_csv("dataset/dataset.csv")

map_kemudahan = {0: "TidakBerlaku", 1: "SangatMudah", 2: "Mudah", 3: "Sulit", 4: "SangatSulit"}


# Cari semua kolom yang ada kata 'kemudahan' di nama kolomnya
kemudahan_cols = [col for col in df.columns if "kemudahan_mencapai" in col.lower()]

# Apply mapping ke semua kolom tsb
df[kemudahan_cols] = df[kemudahan_cols].apply(lambda col: col.map(map_kemudahan))

df.to_csv("dataset_r7_clean.csv", index=False)
print(f"{len(kemudahan_cols)} kolom 'kemudahan' berhasil di-decode ✅")
```

```python
map_topografi = {1: "Tebing", 2: "Lereng", 3: "Dataran", 4:"Lembah"}

topografi_col = [col for col in df.columns if col.startswith("topografi_")]

df[topografi_col] = df[topografi_col].apply(lambda col: col.map(map_topografi))

df.to_csv("dataset_r7_clean.csv", index=False)
print(f"{len(topografi_col)} kolom 'topografi' berhasil di-decode ✅")
```
